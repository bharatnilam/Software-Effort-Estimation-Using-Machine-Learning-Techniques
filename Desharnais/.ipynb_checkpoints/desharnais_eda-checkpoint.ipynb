```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.read_csv('desharnais_clean.csv')
```

```python
ds.corr()
```

```python
colormap = plt.cm.viridis
plt.figure(figsize=(10,10))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.set(font_scale=1.05)
sns.heatmap(ds.astype(float).corr(),linewidths=0.1,vmax=1.0, square=True,cmap=colormap, linecolor='white', annot=True)
```

```python
pd.plotting.scatter_matrix(ds, alpha=0.75, diagonal='kde', figsize=(15,15));
```

```python
ds[['Length', 'Effort']]
```

```python
ds.plot(x = 'Length', y = 'Effort', kind = 'scatter', figsize = (10, 10))
```

```python
pd.set_option('display.max_rows', None)
```

```python
pd.pivot_table(ds, index = ['Length', 'Transactions', 'Entities', 'PointsAdjust', 'PointsNonAjust'], values = 'Effort').sort_values('Effort', ascending = False)
```

```python
ds_pivot = ds[['Length', 'Transactions', 'Entities', 'PointsAdjust', 'PointsNonAjust', 'Effort']]
```

```python
for i in ds_pivot.columns:
    print(i)
    print(pd.pivot_table(ds_pivot, index = i, values = 'Effort').sort_values('Effort', ascending = False))
```

```python
ds_barplot = ds[['Length', 'Transactions', 'Entities', 'PointsAdjust', 'PointsNonAjust', 'Effort']]
```

```python
for i in ds_barplot.columns:
    bar_num = ds_barplot[i].value_counts()
    chart = sns.barplot(x = bar_num.index, y = bar_num)
    chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
    plt.show()
```

```python
for i in ds_barplot.columns:
    ds_barplot.plot(kind='scatter', x=i, y='Effort')
```
